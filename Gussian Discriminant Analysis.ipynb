```python
def phi(result_vector):
    return np.mean(result_vector == 1)
```

```python

def mew_not(X, Y):
    X_class0 = X[Y == 0, :]
    mu0 = np.mean(X_class0, axis=0)
    
    return mu0
```

```python
def mew_one(X, Y):
    X_class1 = X[Y == 1, :]
    mu1 = np.mean(X_class1, axis=0)
    
    return mu1
```

```python
def mew_two(X, Y):
    X_class2 = X[Y == 2, :]
    mu2 = np.mean(X_class2, axis=0)
    
    return mu2
```

```python
mew_two(np.array(feature_vector_iris).T, np.array(result_vector_iris[0]))
```

```python
def sigma(X, Y, mu0, mu1, mu2):
    m, n = X.shape
    sigma = np.zeros((n, n))

    for i in range(m):
        xi = X[i, :]
        if Y[i] == 0:
            diff = (xi - mu0).reshape(n, 1) 
        elif Y[i] == 1:
            diff = (xi - mu1).reshape(n, 1)
        else:
            diff = (xi - mu2).reshape(n, 1)
        sigma += np.dot(diff, diff.T)

    return sigma / m 
```

```python
phi_val = phi(np.array(result_vector_iris[0]))
feature_vector_np = np.array(feature_vector).T
feature_vector_iris_np = np.array(feature_vector_iris).T
mew_not_value = mew_not(feature_vector_iris_np, np.array(result_vector_iris[0]))
mew_one_value = mew_one(feature_vector_iris_np, np.array(result_vector_iris[0]))
mew_two_value = mew_one(feature_vector_iris_np, np.array(result_vector_iris[0]))
sigma_val = sigma(feature_vector_iris_np, np.array(result_vector_iris[0]), mew_not_value, mew_one_value)
print(phi_val)
print(mew_not_value)
print(mew_one_value)
print(sigma_val)
```

```python
def gda_predict(X, mu0, mu1, mu2, sigma, phi):
    m, n = X.shape
    predictions = np.zeros(m)

    sigma_inv = np.linalg.inv(sigma)
    sigma_det = np.linalg.det(sigma)

    constant = 1 / ((2 * np.pi) ** (n / 2) * np.sqrt(sigma_det))

    for i in range(m):
        xi = X[i, :]
        diff0 = xi - mu0
        diff1 = xi - mu1
        diff2 = xi - mu2

        exponent0 = -0.5 * np.dot(np.dot(diff0.T, sigma_inv), diff0)
        exponent1 = -0.5 * np.dot(np.dot(diff1.T, sigma_inv), diff1)
        exponent2 = -0.5 * np.dot(np.dot(diff2.T, sigma_inv), diff2)

        p_x_given_0 = constant * np.exp(exponent0) * (1 - phi)
        p_x_given_1 = constant * np.exp(exponent1) * phi
        p_x_given_2 = constant * np.exp(exponent2) * phi

        p_0_given_x = p_x_given_0
        p_1_given_x = p_x_given_1
        p_2_given_x = p_x_given_2

        if p_0_given_x > p_1_given_x and p_0_given_x > p_2_given_x:
            predictions[i] = 0
        elif p_1_given_x > p_2_given_x and p_1_given_x > p_0_given_x:
            predictions[i] = 1
        else:
            predictions[i] = 2

    return predictions
```

```python
predictions = gda_predict(feature_vector_iris_np, mew_not_value, mew_one_value, mew_two_value ,sigma_val, phi_val)
```
