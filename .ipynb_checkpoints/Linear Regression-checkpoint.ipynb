```python
feature_vector = [data_frame['YearsExperience'].tolist()]
result_vector = [data_frame['Salary'].tolist()]
```

```python
def hypothesis(theta_vector , features , index):
    hyp_index = theta_vector[0]
    for i in range(0 , len(features)) :
        hyp_index += features[i][index] * theta_vector[i+1]
    
    return hyp_index
```

```python
def theta_not(initial_val , result_vector , feature_vector , theta_vector ,  learning_rate):
    error = 0
    for index in range(len(result_vector[0])):
        error += (hypothesis(theta_vector , feature_vector , index) - result_vector[0][index])
    new_theta_not = initial_val - learning_rate * error
    return new_theta_not
```

```python
def theta_feature(initial_val ,feature_index , result_vector , feature_vector , theta_vector , learning_rate):
    error = 0
    for index in range(len(result_vector[0])) :
        error += (hypothesis(theta_vector , feature_vector , index) - result_vector[0][index]) * feature_vector[feature_index][index]
    new_theta_feature = initial_val - learning_rate * error
    return new_theta_feature
```

```python
initial_theta_exp = 5
initial_theta_not = 5
learning_rate = 0.001
cycles = 1000
```

```python
for _ in range(cycles):
    current_theta = [initial_theta_not, initial_theta_exp]
    initial_theta_not = theta_not(current_theta[0], result_vector, feature_vector, current_theta, learning_rate)
    initial_theta_exp = theta_feature(current_theta[1], 0 , result_vector, feature_vector, current_theta, learning_rate)
```

```python
initial_theta_exp
```

```python
initial_theta_not
```

```python
result = []

for i in range(len(result_vector[0])) :
    result.append(hypothesis((initial_theta_not,initial_theta_exp) , feature_vector , i))
```

```python
plt.scatter(data_frame["YearsExperience"] , result)
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

```python

# Load dataset
data_frame = pd.read_csv("/Users/sriramkurnella/AI/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 4 - Simple Linear Regression/Python/Salary_Data.csv")

# Convert to list
x_train = data_frame['YearsExperience'].tolist()
y_train = data_frame['Salary'].tolist()

def knn_predict(x_query, x_train, y_train, k):
    # Step 1: Calculate distances
    distances = []
    for i in range(len(x_train)):
        d = abs(x_query - x_train[i])  # 1D absolute distance
        distances.append((d, y_train[i]))
    
    # Step 2: Sort by distance
    distances.sort(key=lambda tup: tup[0])
    
    # Step 3: Average of k-nearest y values
    neighbors = distances[:k]
    y_pred = sum([val[1] for val in neighbors]) / k
    return y_pred

# Predict across training data for visualization
k = 3
x_test_range = np.linspace(min(x_train), max(x_train), 100)
y_preds = [knn_predict(x, x_train, y_train, k) for x in x_test_range]

# Plot results
plt.scatter(x_train, y_train, color='blue', label='Training Data')
plt.plot(x_test_range, y_preds, color='red', label=f'KNN Prediction (k={k})')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('KNN Regression')
plt.legend()
plt.grid()
plt.show()
```
